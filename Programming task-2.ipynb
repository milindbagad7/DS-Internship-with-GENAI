#### Defanging an IP Address

```python
class Solution:
    def defangIPaddr(self, addr: str) -> str:
        result = ""
        for ch in addr:
            if ch == ".":
                result += "[.]"
            else:
                result += ch
        return result

addr = "1.1.1.1"
solution = Solution()
print(solution.defangIPaddr(addr))
```

#### Find Numbers with Even Number of Digits:


```python
class Solution:
    def findNumbers(self, num):
        count_ = 0
        
        for n in num:
            if len(str(n)) % 2 == 0:
                count_ += 1
                
        return count_

num = [12, 345, 2, 6, 7896]
solution = Solution()
print(solution.findNumbers(num))
```

#### Number of Good Pairs

```python
class Solution:
    def numIdenticalPairs(self, n: list[int]) -> int:
        count_ = 0
        for i in range(len(n)):
            for j in range(i + 1, len(n)):
                if n[i] == n[j]:
                    count_ += 1
        return count_

solution = Solution()

# 3 test cases
print(solution.numIdenticalPairs([1, 2, 3, 1, 1, 3]))
print(solution.numIdenticalPairs([1, 1, 1, 1]))
print(solution.numIdenticalPairs([1, 2, 3, 4]))
```

#### How Many Numbers Are Smaller Than the Current Number

```python
class Solution:
    def smallerNumbersThanCurrent(self, n: list[int]) -> list[int]:
        result = []
        
        for i in range(len(n)):
            count_ = 0
            for j in range(len(n)):
                if n[j] < n[i]:
                    count_ += 1
            result.append(count_)
            
        return result

solution = Solution()

print(solution.smallerNumbersThanCurrent([8, 1, 2, 2, 3]))
print(solution.smallerNumbersThanCurrent([6, 5, 4, 8]))
print(solution.smallerNumbersThanCurrent([7, 7, 7, 7]))
```

#### Subtract the Product and Sum of Digits of an Integer

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        pro = 1
        tot_sum = 0

        while n > 0:
            digi = n % 10
            pro *= digi
            tot_sum += digi
            n //= 10

        return pro - tot_sum

solution = Solution()


print(solution.subtractProductAndSum(234))
print(solution.subtractProductAndSum(4421))
```

#### XOR Operation in an Array

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        res_ = 0
        
        for i in range(n):
            val = start + 2 * i
            res_ ^= val
        
        return res_

solution = Solution()

print(solution.xorOperation(5, 0))
print(solution.xorOperation(4, 3))
```
